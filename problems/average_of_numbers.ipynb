```python
# Logic
# Average = (sum of all numbers) รท (total count)
```

```python
# use in-built function
numbers = [10, 20, 30, 40, 50]
average = sum(numbers) / len(numbers) if numbers else 0
print("Average:", average)
    
```

```python
# Take input from user as a comma-separated string
user_input = input("Enter numbers separated by commas: ")  # e.g., 10,20,30,40

# Convert input string into a list of numbers
# strip() removes any extra spaces before or after the number
# x.strip() checks if the string is not empty
numbers = [float(x.strip()) for x in user_input.split(",") if x.strip()]

# Calculate average
if numbers:
    average = sum(numbers) / len(numbers)
    print("Average:", average)
else:
    print("No numbers entered.")
```

```python
# Average function
def calculate_average(list):
    sum=0
    for i in list:
        sum=sum+i
    average = sum / len(list)
    return average

user_input = input("Enter numbers separated by commas: ")  # e.g., 10,20,30,40
numbers_list = [float(x.strip()) for x in user_input.split(",") if x.strip()]

print ("Average: ", calculate_average(numbers_list))
```
