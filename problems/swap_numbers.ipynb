```python
# Swap 2 number without using 3rd variable
```

Logic - 1 - Swap using python simple way
- a,b = b,a
- or function taking input a,b return b,a

```python
a = int(input("enter a value: "))
b = int(input("enter b value: "))

a,b=b,a
print("a: ", a , "\nb: ", b)
```

```python
def swap(a,b):
    return b,a

a = int(input("enter a value: "))
b = int(input("enter b value: "))

a,b=swap(a,b)
print("a: ", a , "\nb: ", b)
```


Logic - 2 - Using Addition & Subtraction
a = a + b
b = a - b
a = a - b



```python
def swap(a,b):
    a = a + b
    b = a - b
    a = a - b
    return a,b

a = int(input("enter a value: "))
b = int(input("enter b value: "))

a,b=swap(a,b)
print("a: ", a , "\nb: ", b)
```

Logic - 3 - Using Multiplication & Division
a = a * b
b = a / b
a = a / b

```python
def swap(a,b):
    a = a * b
    b = a / b
    a = a / b
    return a,b

a = int(input("enter a value: "))
b = int(input("enter b value: "))

a,b=swap(a,b)
print("a: ", a , "\nb: ", b)
```
